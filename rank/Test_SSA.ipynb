```julia
using PyPlot
using FFTW
using LinearAlgebra
using SeisReconstruction
using HDF5
using SeisPlot, SeisProcessing, SeisReconstruction


include("SSA.jl")
include("rqrd.jl")
```

```julia
d=SeisLinearEvents();
dn = SeisAddNoise(d,.9)
p = 2
df = SSA(dn,0.004, p, .4, 80.0, "rqrd")

c = 1*maximum(abs.(dn))

figure(1);
subplot(131);SeisPlotTX(d, cmap="seismic", pclip=100,vmin=-c,vmax=c,fignum=1,title="True")
subplot(132);SeisPlotTX(dn,cmap="seismic",pclip=100,vmin=-c,vmax=c,fignum=1,title="Noisy")
subplot(133);SeisPlotTX(df,cmap="seismic",pclip=100,vmin=-c,vmax=c,fignum=1,title="SSA p=2")

nt,nx=size(d)
tight_layout()
savefig("ssa_2d_1.png")

figure(2);
K = collect(2:10:nx)
subplot(131);SeisPlotTX( d[:,K], style="wiggles", cmap="seismic", pclip=100,vmin=-c,vmax=c,fignum=2,title="True")
subplot(132);SeisPlotTX(dn[:,K], style="wiggles", cmap="seismic", pclip=100,vmin=-c,vmax=c,fignum=2,title="Noisy")
subplot(133);SeisPlotTX(df[:,K], style="wiggles", cmap="seismic", pclip=100,vmin=-c,vmax=c,fignum=2,title="SSA p=2")
tight_layout()
savefig("ssa_2d_2.png")
```
