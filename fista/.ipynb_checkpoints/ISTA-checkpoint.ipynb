```julia
"""
    Estimation of sparse solution x of the proble
    x = argmin ||A x- y||_2^2 + λ ||x||_1  
"""

function ISTA(A,y,Niter,λ)

    
    Soft(x,alpha) = sign(x)*max(abs(x)-alpha, 0)

    N,M = size(A)
    e = Power_Iteration(A)
    η = 0.95/e
   
  println(size(y))
    x = zeros(Float64,M)
  
    J = zeros(Niter)
    for k = 1:Niter
        u = x .-  η*A'*(A*x.-y)
        x = Soft.(u, η*λ)
        J[k] = 0.5*sum((A*x-y).^2) + λ*sum(abs.(x))
    end
        return x, J
end

    
"""
    The max eigvalue of A'A which is also the max eigenvalue of AA' 
    is estimated via the Power Iteration Method
"""
function Power_Iteration(A)

    
    N,M = size(A)
    
    if N > M 
        H = A'*A;
        b = randn(M,1)
    else
        H = A*A';
        b = randn(N,1)
    end
        e = 1.
        
        for k = 1:20
            tmp = H*b
            e = norm(tmp)
            tmp = tmp/e
            b = tmp
        end
        
        return e
    end

```

```julia
 N = 90
 M = 150
 A = randn(N,M)
 
 
 xt = Array(sprandn(M,1,0.171))
  y = A*xt 


  x,J = ISTA(A,y,1400,0.91)

subplot(221); plot(xt); title("True sparse model")
subplot(222); plot(y);  title("Data")
subplot(223); plot(x);  title("Recovered sparse model")
subplot(224); plot(J);  title("ISTA convergence curve")
tight_layout()

rmse = sum((xt-x).^2)/sum(xt.^2)

@printf "rmse %0.8e  \n" rmse;
```
